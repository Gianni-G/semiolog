```python
import os
os.chdir("../")

import semiolog as slg

semiotic = slg.Cenematic("en_bnc_test")
```

```python
corpus_str = semiotic.corpus.dev.map(lambda sent: {"text": semiotic.vocab.normalizer(sent["text"])})
corpus_str = "".join(corpus_str["text"])
```
